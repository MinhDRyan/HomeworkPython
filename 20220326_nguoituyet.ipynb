```python
import turtle 
tl = turtle.Turtle()
tl1 = turtle.Turtle()
tl2 = turtle.Turtle()
#tl.hideturtle()
tl1.hideturtle()
tl2.hideturtle()


screen = tl.getscreen()
screen.bgcolor('#87CEFA')
tl.speed(0)
y=100
x=7*y/10

tl.fillcolor("white")

tl.penup()
tl.setheading(-90)
tl.fd(y)
tl.setheading(0)
tl.pendown()
tl.begin_fill()
tl.circle(y)
tl.end_fill()

tl1.pensize(y/10)
tl1.dot()
tl1.penup()
tl1.setheading(90)
tl1.forward(y/3)
tl1.dot()
tl1.pendown()
tl1.penup()
tl1.setheading(270)
tl1.forward(2*y/3)
tl1.dot()
tl1.pendown()


tl2.penup()
tl2.setheading(90)
tl2.forward(y)
tl2.pendown()

tl2.fillcolor("red")
tl2.begin_fill()
tl2.setheading(270)
tl2.circle(y,90)
tl2.setheading(180)
tl2.circle(y,-30)

tl2.setheading(210)
tl2.circle(y,-20)
tl2.setheading(30)
tl2.circle(y,20)

tl2.setheading(160)
tl2.circle(2*y,30)
tl2.setheading(-5)
tl2.circle(2*y,-25)
tl2.end_fill()


tl.penup()
tl.setheading(90)
tl.fd(2*y-x/2)
tl.setheading(0)
tl.pendown()
tl.begin_fill()
tl.circle(x)
tl.end_fill()

tl.fillcolor("red")
tl.begin_fill()
tl.circle(x,-50)
tl.right(180)
tl.circle(x/5,180)
tl.circle(7*x/5,100)
tl.circle(x/5,180)
tl.right(180)
tl.circle(x,-50)
tl.end_fill()

tl.setheading(0)
tl.penup()
tl.circle(x,135)
tl.pendown()
tl.setheading(180)
tl.fillcolor("red")
tl.begin_fill()
tl.forward(2**0.5*x)
tl.setheading(270)
tl.circle(2**0.5*x/2,-180)
tl.end_fill()
tl.setheading(180)
tl.pensize(x/4)
tl.pencolor("#EE0000")
tl.forward(2**0.5*x)
tl.penup()
tl.setheading(270)
tl.circle(2**0.5*x/2,-90)
tl.pendown()
tl.pensize(x/6)
tl.dot("#EE0000")

tl2.penup()
tl2.home()
tl2.setheading(90)
tl2.fd(y-x/2 +x)
tl2.pendown()
tl2.pensize(x/7)
tl2.dot("#FF0000")
tl2.pensize(x/10)
tl2.dot("#FF4500")


tl2.penup()
tl2.fd(x/3)
tl2.setheading(0)
tl2.fd(x/3)
tl2.pendown()
tl2.pensize(x/10)
tl2.dot()
tl2.penup()
tl2.setheading(180)
tl2.fd(2*x/3)
tl2.pendown()
tl2.pensize(x/10)
tl2.dot()

tl2.penup()
tl2.setheading(0)
tl2.fd(x/3)
tl2.setheading(270)
tl2.fd(x)
tl2.setheading(0)
tl2.pendown()
tl2.pensize(2)
tl2.circle(x,45)
tl2.circle(x,-90)






turtle.done()
```
